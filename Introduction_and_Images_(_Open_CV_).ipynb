```python
import cv2
from google.colab.patches import cv2_imshow
import numpy as np


img = cv2.imread('/content/download.jfif',1)


if img is not None:

    imgBlue = img[:,:,0]
    imgGreen = img[:,:,1]
    imgRed = img[:,:,2]

    new_img = np.hstack((imgBlue, imgGreen, imgRed))
    cv2_imshow(new_img)

else:
    print("Error loading image.")
```
