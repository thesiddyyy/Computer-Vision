```python
import cv2
import numpy as np
```

```python
img = np.zeros((512,512,3))

from google.colab.patches import cv2_imshow

cv2_imshow(img)
```

```python
img = np.zeros((512,512,3))


from google.colab.patches import cv2_imshow


cv2_imshow(img)
```

```python
cv2.rectangle(img,pt1=(100,100),pt2=(300,300),color=(0,0,255),thickness=3)
cv2_imshow(img)
```

```python
cv2.rectangle(img,pt1=(100,100),pt2=(300,300),color=(0,0,255),thickness=-1)
cv2_imshow(img)
```

```python
cv2.circle(img,center=(100,100),radius=40,color=(0,255,0),thickness=6)
cv2_imshow(img)
```

```python
cv2.line(img,pt1=(0,0),pt2=(512,512),color=(255,0,0),thickness=3)
cv2_imshow(img)
```

```python
cv2.putText(img,org=(50,50), fontScale=4,color=(0,255,255),thickness=2,lineType=cv2.LINE_AA,text="Namaste",fontFace=cv2.FONT_ITALIC)
cv2_imshow(img)
```
